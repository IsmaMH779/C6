```python
import matplotlib.pyplot as plt
```

```python
#1

categorias = ['A','B','C','D']
valores = [15, 7, 10, 12]

fig, ax = plt.subplots()
ax.bar(categorias, valores)

plt.show()
```

```python
#2
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

fig2, ax = plt.subplots()
ax.plot(x,y)

plt.show()
```

```python
#3

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

fig3, ax = plt.subplots()
plt.scatter(x,y)
plt.show()
```

```python
#4

porcentajes = [30, 20, 25, 15, 10]
etiquetas = ['A', 'B', 'C', 'D', 'E']

fig4, ax = plt.subplots()
ax.pie(porcentajes,labels = etiquetas)
plt.show()
```
